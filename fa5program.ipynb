```python
def calculate_total_price(price, quantity) :
   total = price * quantity
   return total

user_price = float(input("Enter the price of the item: "))
user_quantity = int(input("Enter the quantity of the item: "))

final_total = calculate_total_price(user_price, user_quantity)
print(f"The total price for {user_quantity} items is: {final_total}")
```
